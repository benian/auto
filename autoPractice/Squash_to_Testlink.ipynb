```python
from xml.etree import ElementTree
from xml.dom import minidom

def prettify(elem):
    """Return a pretty-printed XML string for the Element.
    """
    rough_string = ElementTree.tostring(elem, 'utf-8')
    reparsed = minidom.parseString(rough_string)
    return reparsed.toprettyxml(indent="  ")

def genxml(df):
    suitename = df["Directory"].values[0]
    filename = "data/"+ suitename + ".xml"
    toptestsuite = Element('testsuite')
    toptestsuite.set("name",suitename)

    former_id = 0
    subsuitename = ""
    testcasename = ""
    for index, row in df.iterrows():
        if len(row["PATH"].split("/")) <=2:
            #print(row["PATH"])
            subtestsuite = toptestsuite  
        elif row["PATH"].split("/")[2] != subsuitename:
            subsuitename = row["PATH"].split("/")[2]
            subtestsuite = SubElement(toptestsuite, 'testsuite')
            subtestsuite.set("name",subsuitename)
        if row["LABEL"]!= testcasename:
            testcasename = row["LABEL"]
        if row["ID"] != former_id: 
            count = 1
            #new test case
            testcase = SubElement(subtestsuite, 'testcase')
            testcase.set("name",testcasename)
            summary = SubElement(testcase, 'summary')
            steps = SubElement(testcase, 'steps')
            former_id = row["ID"]

        step = SubElement(steps, 'step')
        step_number = SubElement(step, 'step_number')
        step_number.text  = str(count)
        actions = SubElement(step, 'actions')
        actions.text = row["ACTION"]
        expectedresults = SubElement(step, 'expecteresults')
        count +=1
    #print(prettify(toptestsuite))    
    with open(filename, 'w', errors="ignore") as file:
        file.write(prettify(toptestsuite))
```
